```python
# Using HSV - for color filtering
# H-> 0-180,  S-> 0-255, V-> 0-255
input = cv2.imread('/content/download.jpg')
cv2_imshow(input) 
hsv_image = cv2.cvtColor(input, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_image)
```

```python
# Ampliying the image of blue color
B,G,R = cv2.split(input)
cv2.waitKey(0)
cv2.destroyAllWindows()
merged_input = cv2.merge([B+100,G,R])
```

```python
cv2_imshow(merged_input)
```

```python
# Ampliying the image of green color
B,G,R = cv2.split(input)
cv2.waitKey(0)
cv2.destroyAllWindows()
merged_input = cv2.merge([B,G+100,R])
cv2_imshow(merged_input)
```

```python
# Ampliying the image of red color
B,G,R = cv2.split(input)
cv2.waitKey(0)
cv2.destroyAllWindows()
merged_input = cv2.merge([B,G,R+100])
cv2_imshow(merged_input)
```

```python
# Ampliying the image of all colors
B,G,R = cv2.split(input)
cv2.waitKey(0)
cv2.destroyAllWindows()
merged_input = cv2.merge([B,G,R])
cv2_imshow(merged_input)
```

```python
# Creating a matrix of zeros with dimension of imaage h x w
B,G,R = cv2.split(input)
zeros = np.zeros(input.shape[:2], dtype = "uint8")
cv2_imshow(cv2.merge([zeros,zeros,R]))
cv2_imshow(cv2.merge([zeros,G,zeros]))
cv2_imshow(cv2.merge([B,zeros,zeros]))
```

## **Histograms**


```python
from matplotlib import pyplot as plt
histogram = cv2.calcHist([input],[0],None,[256],[0,256])
plt.hist(input.ravel(),256,[0,256]);plt.show()  # ravel -> takes 2d converts into 1d
# viewing separate color channels
color = ('b','g','r')
# Now separate colors and plot each in histogram
for i, col in enumerate(color):
  histogram2 = cv2.calcHist([input],[i],None,[256],[0,256])
  plt.plot(histogram2,color = col)
  plt.xlim([0,256])
plt.show()
```

## **Drawing Images and shapes**

```python
# Making a black square
#creating a black image
image = np.zeros((512,512,3), np.uint8)
# creating in black and white
image_bw =  np.zeros((512,512), np.uint8)

# Displaying them
cv2_imshow(image)

```

```python
cv2_imshow(image_bw)
```

```python
# Lets draw a line over the black square
# draw a diagonal blue line of thickness 5 pixels
image = np.zeros((512,512,3), np.uint8)
# cv2.line(image, starting coordinates,ending coordinates,color,thickness)
cv2.line(image, (0,0), (511,511),(255,127,0),5)
cv2_imshow(image)
```

```python
# Drawing a rectangle
image = np.zeros((512,512,3), np.uint8)
# cv2.rectangle(image, starting vertex,oposite vertex,color,thickness)
cv2.rectangle(image, (100,100), (300,250),(127,50,127),5)
cv2_imshow(image)
```

```python
# Drawing a circle
image = np.zeros((512,512,3), np.uint8)
# cv2.circle(image, center,radius,color,fill)
cv2.circle(image, (350,350), 100,(15,75,50),-1)
cv2_imshow(image)
cv2.destroyAllWindows()
```

```python
# Creating polygons
# defining 4 points
pts =np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)

# reshaping in form required by polylines
pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts],True, (0,0,255), 3)
cv2_imshow(image)
```

```python
# Adding text with cv2.putText
# putText(img, text_to_display, bottom_left_starting_point,fontFace, fontSize, color, thickness) 
image = np.zeros((512,512,3), np.uint8)
cv2.putText(image, "Hello",(75,290),cv2.FONT_HERSHEY_COMPLEX, 2, (100,170,0),3) 
cv2_imshow(image)
```
